### First model select

Logistic regression and random forest look promising.

Next step is tweaking hyper parameters and the threshold to maximize F1 score on my selected models.

#### Manual tweaking, Logistic Regression

```python
clf = LogisticRegression(C=1, penalty='l1', class_weight={1:1, 0:1}, solver='liblinear')
cross_val_model(X_train_val, y_train_val, clf, scale=True, nr_splits=5, fold_type='stratified', threshold=0.3, over_sample=1)
```

```python
clf.fit(X_train_val, y_train_val)
precision_curve, recall_curve, threshold_curve = precision_recall_curve(y_test, clf.predict_proba(X_test)[:,1] )

plt.figure(dpi=80)
plt.plot(threshold_curve, precision_curve[1:],label='precision')
plt.plot(threshold_curve, recall_curve[1:], label='recall')
plt.legend(loc='lower left')
plt.xlabel('Threshold');
plt.title('Precision and Recall Curves');
```

```python
fpr, tpr, thresholds = roc_curve(y_train_val, clf.predict_proba(X_train_val)[:,1])
plt.plot(fpr, tpr,lw=2)
plt.plot([0,1],[0,1],c='violet',ls='--')
plt.xlim([-0.05,1.05])
plt.ylim([-0.05,1.05])


plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve for student pass/fail');
```

```python
# Coefficients
importance = [abs(val) for val in clf.coef_[0]]
indices = np.argsort(importance)[::-1]
for f in range(X_train_val.shape[1]):
    print(f'{f+1}. {X_train_val.columns[indices[f]]:<20} | {importance[indices[f]]:.3f}')
```

For logistic regression I can examine the coefficients, but this will not be possible when using random forest or an ensemble. To still get some interpretation I will also compare the feature importance generated by permutation importance.

```python
#Feature importance with permutation feature importance
from sklearn.inspection import permutation_importance
r = permutation_importance(clf, X_train_val, y_train_val, scoring = 'f1',
                            n_repeats=30, random_state=0)

importances = r.importances_mean
std = r.importances_std
indices = np.argsort(importances)[::-1]
for f in range(X_train_val.shape[1]):
    print(f'{f+1}. {X_train_val.columns[indices[f]]:<20} | {importances[indices[f]]:.3f}')
```

#### Grid Search with F1 scoring

Sticking with logistic regression and random forest, I tried a grid search to see if I could find a combination of parameters that works better than my manual tweaks above

```python
from sklearn.model_selection import GridSearchCV
lr_param = {'C': [0.001, 0.01, 0.1, 0.5, 1, 2, 5, 10, 100, 1000], 'penalty': ['l1', 'l2']}
search_lr = GridSearchCV(LogisticRegression(solver='liblinear'), lr_param, scoring='roc_auc', n_jobs=-1, verbose=1)
search_lr.fit(X_train_val, y_train_val)
print(search_lr.best_estimator_, search_lr.best_score_)
```

```python
lr = LogisticRegression(C=1, penalty='l1', class_weight={1:1, 0:1}, solver='liblinear')
cross_val_model(X_train_val, y_train_val, lr, scale=True, nr_splits=5, fold_type='stratified', threshold=0.3, over_sample=1)
```

```python
rf_param = {'criterion': ['gini', 'entropy'],
            'n_estimators': [50, 100, 150, 200, 300],
            'max_features': ['auto', 'log2'],
            'max_depth': [4, 8, 12, 16]}
search_rf = GridSearchCV(RandomForestClassifier(), rf_param, scoring='f1', n_jobs=-1, verbose=0)
search_rf.fit(X_train_val, y_train_val)
print(search_rf.best_estimator_, search_rf.best_score_)
```

```python
rf = RandomForestClassifier(criterion='entropy', max_depth=6, max_features='log2',min_samples_leaf=1, n_estimators=500, class_weight={1:1, 0:1})
cross_val_model(X_train_val, y_train_val, rf, scale=True, nr_splits=5, fold_type='stratified', threshold=0.3, over_sample=1)
```

For both of these models, the Gridsearch does not take into account the threshold and my manually fitted models perform better (F1 score) than the outcome of the Gridsearch.

### Final model

```python
lr.fit(X_train_val, y_train_val)
rf.fit(X_train_val, y_train_val)
voting_classifer = VotingClassifier(
            estimators=[('rf', rf), ('lr', lr)],
            voting='soft',
            n_jobs=-1)
ens = voting_classifer.fit(X_train_val, y_train_val)
threshold = 0.3
```

```python
from sklearn.inspection import permutation_importance
r = permutation_importance(ens, X_test, y_test, scoring = 'f1',
                            n_repeats=30, random_state=0)

importances = r.importances_mean
std = r.importances_std
indices = np.argsort(importances)[::-1]
for f in range(X_test.shape[1]):
    print(f'{f+1}. {X_test.columns[indices[f]]:<20} | {importances[indices[f]]:.3f}')
```

```python
y_score_test = ens.predict_proba(X_test)[:, 1]
y_pred_test = (y_score_test > threshold)
acc = accuracy_score(y_test, y_pred_test)
roc_auc = roc_auc_score(y_test, y_score_test)
f1_test = f1_score(y_test, y_pred_test)

print(f'acc: {acc:.3f} | roc_auc: {roc_auc:.3f} | f1: {f1_test:.3f}')
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_test)
plt.figure(dpi=150)
sns.heatmap(conf_matrix, cmap=plt.cm.Blues, annot=True, square=True)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Final confusion matrix');
```
